```python
# Initial imports.
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load Data
loans_df = pd.read_csv('loans_df.csv', index_col = 'LoanNumber' )
loans_df
```

# Prep to Encode Data

```python
# List all columns
loans_df.info()
```

```python
# Date - Split into month, day, year 
loans_df[['MonthApproved','DayApproved','YearApproved']] = loans_df.DateApproved.str.split('/',expand=True)
loans_df[['MonthApproved','DayApproved','YearApproved']]
```

```python
loans_df['BorrowerName'].duplicated().sum()
```

```python
# 'BusinessAgeDescription' column - drop nulls
loans_df = loans_df.dropna(subset=['BusinessAgeDescription'])
loans_df
```

```python
# 'BusinessAgeDescription' column - drop nulls
loans_df = loans_df.dropna(subset=['ForgivenessAmount'])
loans_df
```

```python
# Drop unneeded columns
loans_df.drop(columns=['Unnamed: 0', 'DateApproved','LoanStatusDate', 'ForgivenessDate','DayApproved','BorrowerAddress',
                       'category_title'], inplace=True)
loans_df.head(10)
```

```python
# Create new df with info
info_df = loans_df[['BorrowerName','Race','Ethnicity','Gender','Veteran']]
info_df
```

```python
loans_df.drop(columns=['BorrowerName','Race','Ethnicity','Gender','Veteran'], inplace=True)
loans_df.head(10)
```

```python
loans_df.info()
```

# Encode Data and save to csv 

```python
# Encode using get_dummies method
encoded_loans_df = pd.get_dummies(loans_df, columns=['BusinessAgeDescription', 'CD','BusinessType',
                                                     'Yelp_Category', 'MonthApproved','YearApproved'])
encoded_loans_df
```

```python
# Save encoded and info dfs to csvs
encoded_loans_df.to_csv('encoded_loans_dfV2.0.csv')
info_df.to_csv('info_df.csv')
```
