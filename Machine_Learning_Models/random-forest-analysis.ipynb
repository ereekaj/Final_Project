```python
# Initial imports.
import pandas as pd
from path import Path
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# Loading data
file_path = Path("Machine Learning CSVs/encoded_loans_df.csv")
df_business = pd.read_csv(file_path, index_col="LoanNumber")
df_business
```

```python
# Encode LoanStatus column
le = LabelEncoder()
df_encode = df_business.copy()
df_encode['LoanStatus'] = le.fit_transform(df_encode['LoanStatus'])
df_encode
```

```python
df_encode.dtypes
```

```python
# looking for null values
df_encode.isna().sum()
```

```python
# drop ForgivenessAmount column
df_encode = df_encode.drop('ForgivenessAmount', axis = 1)
df_encode = df_encode.drop('CurrentApprovalAmount', axis = 1)
df_encode
```

```python
# Define features set
X = df_encode.copy()
X = X.drop("LoanStatus", axis=1)
X
```

```python
# Define target vector
y = df_encode["LoanStatus"].ravel()
y[:5]
```

```python
# Splitting into Train and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Creating StandardScaler instance
scaler = StandardScaler()
```

```python
# Fitting Standard Scaller
X_scaler = scaler.fit(X_train)
```

```python
# Scaling data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Fitting the Random Forest Model

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=100, random_state=1) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

## Making Predictions Using the Tree Model

```python
# Making predictions using the testing data
predictions = rf_model.predict(X_test_scaled)
predictions
```

## Model Evaluation

```python
# Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(
    cm, index=["Actual 0 - Exemption 4", "Actual 1 - Paid in Full"], columns=["Predicted 0 - Exemption 4", "Predicted 1 - Paid in Full"]
)

# Calculating the accuracy score
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

## Rank the Importance of Features

```python
# Calculate feature importance in the Random Forest model.
importances = rf_model.feature_importances_
importances
```

```python
# sort the features by their importance.
sorted(zip(rf_model.feature_importances_, X.columns), reverse=True)
```
