```python
# Import the dependencies.
import pandas as pd
import numpy as np
import requests
import json
import csv
from pprint import pprint
from config import yelp_api_key
```

```python
# Yelp URL
url = 'https://api.yelp.com/v3/businesses/search'
```

```python
# Pull parameters
url_params = {'categories':'restaurants', 'location': '80228','limit': 10}
# url_params = {'categories':'restaurants', 'location': '80216' , 'location': 'Denver','limit': 50}
# url_params = {'categories':'restaurants', 'location': 'Denver' , 'location[zip_code]':'80216', 'offset': 0, 'limit': 50}
# url_params = {'categories':'restaurants', 'location.zip_code': '80216' ,  'offset': 0, 'limit': 50}
```

```python
# Headers
headers = {
         'Authorization': 'Bearer %s' % yelp_api_key,
    }
```

```python
# Pull JSON 
response = requests.request('GET', url, headers=headers, params=url_params)

pprint(response.json())
```

```python
response.json()['businesses'][0]

```

```python
total_businesses = response.json()['businesses']
```

```python
list = []

for biz in total_businesses:
    name = biz['name']
    phone = biz['phone']
    addresses = biz['location']['display_address']
    address = '  '.join(addresses)
    
    services = biz['categories']
    
    alias = ''
    title = ''
    
    for service in services:
        alias += service['alias'] + ', '
        title += service['title'] + ', '
    
    rating = biz['rating']
    review_count = biz['review_count']
    is_closed = biz['is_closed']
    lat = biz['coordinates']['latitude']
    lon = biz['coordinates']['longitude']
    
    if lat is None:
        continue
    if lon is None:
        continue

    yelp_dict = {
                    'name': name,
                    'phone': phone,
                    'address' : address,
                    'category_alias': alias,
                    'category_title': title,
                    'rating' : rating,
                    'review_count' : review_count,
                    'lat': lat,
                    'lon': lon,
                    'is_closed': is_closed
                        }

    list.append(yelp_dict)
    print(biz['coordinates'])
```

```python
pprint(list)
```

```python
restaurant_df = pd.DataFrame.from_dict(list)
restaurant_df
```

```python
restaurant_df.to_csv('Resources/restaurant_df.csv')
```
